```python
cp.shape
```

```python
cp.drop([9, 44, 45, 129], inplace = True) 
```

```python
cp.rename(columns={'make':'Fuel_Type',
                   'aspiration': 'No_of_Doors', 
                    'fuel_type': 'Aspiration',
                     'num_of_doors': 'Body_Style',
                      'body_style': 'Drive_Wheels',
                      'drive_wheels':'Engine_Location',
                      'engine_location':'Wheel_Base'
                      , 'wheel_base': 'Manufacturer'}, inplace=True)
```

```python
cp.head(10)
```

```python
cp.shape
```

```python
cp.isnull().sum()
```

```python
cp.describe(include='all')
```

```python
import numpy as np
cp.replace('?', np.nan, inplace = True)
cp.shape
```

```python
cp.describe(include='all')
```

```python
cp.normalized_losses.value_counts()
cp.peak_rpm.value_counts()
cp.No_of_Doors.value_counts()
cp.bore.value_counts()
cp.stroke.value_counts()
cp.horsepower.value_counts()
cp.isnull().sum()
```

```python
cp.info()
```

```python
cp.normalized_losses  = cp.normalized_losses .fillna('161')
cp.No_of_Doors   = cp.No_of_Doors.fillna('four')
cp.bore  = cp.bore .fillna('3.62')
cp.stroke  = cp.stroke .fillna('3.4')
cp.horsepower   = cp.horsepower  .fillna('68')
cp.peak_rpm   = cp. peak_rpm  .fillna('5500')
cp.isnull().sum()
```

```python
cp['normalized_losses'] = cp['normalized_losses'].astype('int64')
cp['bore'] = cp['bore'].astype('float64')
cp['stroke'] = cp['stroke'].astype('float64')
cp['horsepower'] = cp['horsepower'].astype('int64')
cp['peak_rpm'] = cp['peak_rpm'].astype('int64')
cp['price'] = cp['price'].astype('int64')
```

```python
cp.info()
```

```python
import sklearn
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
cp[cp.select_dtypes(include=['object']).columns]  = cp[cp.select_dtypes(include=['object']).columns].apply(le.fit_transform)
```

```python
cp_x = cp.drop(['price'], axis=1)

cp_y  =  cp.price
```

```python
from sklearn.model_selection import train_test_split

cp_x_train, cp_x_test, cp_y_train, cp_y_test = train_test_split(cp_x , cp_y , test_size = .1 , random_state = 101)
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
reg.fit(cp_x_train, cp_y_train)
```

```python
Rsquare = reg.score(cp_x_train, cp_y_train)
print('value of Rsquare is --->', Rsquare)
```

```python
cp_x_train.shape
```

```python
K = 25
N = 180
AdjRsquare = 1 - (1-Rsquare)* (N - 1) / (N - K - 1)
print('value of AdjRsquare is ---> ', AdjRsquare)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
coef_values = pd.DataFrame({'Feature Name' : cp_x_train.columns , 'Coeff' : reg.coef_})
coef_values
coef_values.sort_values('Coeff')
```

```python
pred_train = reg.predict(cp_x_train)
error_train = cp_y_train - pred_train
error_train
import matplotlib.pyplot as plt
plt.hist(error_train)
```

```python
import seaborn as sns
sns.distplot(error_train)
```

```python
plt.plot(error_train)
plt.axhline(y = 0 , linestyle = "--", color = 'red') 
plt.axhline(y =5000, linestyle = "--", color = 'b')
plt.axhline(y = -5000 , linestyle = "--", color = 'g')
```

```python
pred_test = reg.predict(cp_x_test)
pred_test
```

```python
error = cp_y_test - pred_test
error
```

```python
import numpy as np
MSE = np.mean(error*error)
MSE
```

```python
RMSE = np.power(MSE, .5)
RMSE
```

```python
MAPE = np.mean( np.absolute(error / cp_y_test) ) * 100
MAPE
```

```python
accu = 100 - MAPE
accu
```
